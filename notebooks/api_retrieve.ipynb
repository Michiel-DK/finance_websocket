```python
import requests
import os
import pandas as pd
```

```python
params = {
    'apikey' : os.getenv('FINANCE_API_KEY'),
}

url = 'https://financialmodelingprep.com/api/v3/historical-price-full/BTCUSD?'
```

```python
response = requests.get(url, params=params).json()
```

```python
df = pd.DataFrame.from_records(response)
```
